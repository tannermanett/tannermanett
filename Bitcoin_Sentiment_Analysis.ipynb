```python
import pandas as pd
from textblob import TextBlob
import numpy as np
import re
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

df = pd.read_csv (r'/Users/tmane/OneDrive/Desktop/DataAnalysisCourse/Bitcoin_tweets.csv')
print (df)
```

```python
btc_df = pd.DataFrame(df, columns=['text', 'date', 'user_followers', 'user_name'])


btc_df[btc_df['date'] > '2022-01-02 23:59:04']
```

Get Tweets

```python
def cleanTwt(twt):
    twt = re.sub('#bitcoin', 'bitcoin', str(twt))
    twt = re.sub('#Bitcoin', 'Bitcoin', str(twt))
    twt = re.sub('#[A-za-z0-9]+', '',str(twt))
    twt = re.sub('\\n', '',str(twt))
    twt = re.sub('https?:\/\/S+', '', str(twt))
    return twt

```

```python
btc_df['clean_text'] = btc_df['text'].apply(cleanTwt)
btc_df
```

Graph

```python
def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

    

btc_df['Subjectivity'] = btc_df['clean_text'].apply(getSubjectivity)
btc_df['Polarity'] = btc_df['clean_text'].apply(getPolarity)

btc_df
```

```python
#get sentiment text
def getSentiment(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Postive'
```

```python
#Create a column to store text sentiment
btc_df['Sentiment'] = btc_df['Polarity'].apply(getSentiment)
btc_df
```

```python
plt.figure(figsize=(8,6))
for i in range(0, btc_df.shape[0]):
    plt.scatter(btc_df['Polarity'][i], btc_df['Subjectivity'][i], color = 'Purple')
plt.title('Sentiment Analysis Scatter Plot')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```

```python
btc_df['Sentiment'].value_counts().plot(kind='line')
plt.title('Sentiment Analysis Bar Plot')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
```
